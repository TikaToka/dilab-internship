```python
#train
train_new1 = {}
datas = []
brs = "<b>"
bre = "</b>"
br = "<br>"
# start = 0
# end = 3000
for i, data in tqdm(enumerate(train)):
    # if not start <= i < end:
    #     continue
    final = {}
    para = {}
    #paragraph all
    para = {}
    temp = ""
    prevlen = 0
    for j, cont in enumerate(data["context"]):
        for k in range(len(cont[1])):
            temp += brs
            temp += "Sent " + str(prevlen+(k+1)) + ": "
            temp += bre
            temp += cont[1][k]
            temp += br
        prevlen += len(cont[1]) 
    
    text = temp

        #questions
    question = data["question"]
            #sentences_used
    su = []
    for t in range(len(data["supporting_facts"])):
        plen = 0
        for k in range(len(data["context"])):
            if data["supporting_facts"][t][0] == data["context"][k][0]:
                su.append(plen + data["supporting_facts"][t][1])
            else:
                plen += len(data["context"][k][1])
    sentences_used = su
    #answers
    answer = []
    temp = {}
    temp["text"] = data["answer"]
    temp["isAnswer"] = True
    temp["scores"] = {}
    ans = temp
    answer.append(temp)
        #idx
    idx = 0
        #multisent??
    multisent  = True
    qs = {
        "question": question,
        "sentences_used": su,   
        "answers": answer,
        "idx": idx,
        "multisent": multisent
    }

    #id
    id = data["_id"]
    paragraph = {
        "text": text,
        "questions": [qs],
    }
    
    final["paragraph"] = paragraph
    final["id"] = id
    datas.append(final)

train_new1["data"] = datas

with open("hotpot_modded_train.json", "w") as nt:
    json.dump(train_new1, nt1)

print("Done")
```

Build Hotpot structured Dataset with Gold-Sents extractedby AIR

```python
import json

res = open('/content/drive/MyDrive/Yonsei/Graduate/AIR-retriever_onlyDev/MultiRC_justifications_Iterative_Alignment_CLASSIFICATION_parallel_concat_evidence_5/dev.tsv', 'r')
with open("/content/drive/MyDrive/Yonsei/Graduate/hotpot/hotpot_dev_distractor_v1.json") as c:
    ref = json.load(c)
res.readline()
lines = res.readlines()
result = []
for i in lines:
    temp = i.rstrip().split("\t")
    id = temp[1]
    question = temp[3]
    answer = temp[4]
    sens = temp[5].split("  ")
    sup_fact = []
    for j in range(len(sens)):
        sup_fact.append(["title", j])
    for k, d in enumerate(ref):
        if d["_id"] == id:
            ty= d["type"]
            level = d["level"]
    final = {
        "_id": id,
        "answer": answer,
        "question": question,
        "supporting_facts": sup_fact,
        "context": [["title", sens]],
        "type": ty,
        "level": level
    }
    result.append(final)
with open('/content/drive/MyDrive/Yonsei/Graduate/hotpot_dev_distractor_v1', 'w') as fin:
    json.dump(result, fin)
res.close()
```
